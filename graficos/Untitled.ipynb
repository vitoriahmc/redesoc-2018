```python
import pandas as pd
import seaborn as sns; sns.set()
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import os
print(os.getcwd())
```

```python
dados = pd.read_csv('cabreuvalimpa.csv')
```

```python
colunas1 = ['CORDEPELE','FREQUENCIADESENTIREMINSEGUROS'] 
grafico1 = pd.crosstab(dados.FREQUENCIADESENTIREMINSEGUROS, dados.CORDEPELE)
#grafico1.plot(kind='bar', legend=False)
#plt.ylabel('Frequência Absoluta')
sns.heatmap(grafico1, cmap="Greens")
```

```python
colunas2 = ['SEXO','HOMENSSÃOFRACOSOUFORTES'] 
grafico2 = pd.crosstab(dados.HOMENSSÃOFRACOSOUFORTES, dados.SEXO)
grafico2.plot(kind='bar', color = ('red', 'blue'))
plt.ylabel('Frequência Absoluta')
```

```python
idade_jovens_escola = dados.IDADE[(dados.TURMA != '1EJAA') & (dados.TURMA != '1EJAB') & (dados.TURMA != '2EJAA') & (dados.TURMA != '3EJAA')]
grafico3 = pd.crosstab(idade_jovens_escola, dados.MEDODESOFRERBULLYINGNAESCOLA)
grafico3.plot(kind='bar', legend=False)
plt.ylabel('Frequência Absoluta')
```
